The thing that is not this notebook is referred to as "The GUI" for graphical user interface. Technically everything you see is GUI, but here I mean the graph and links to the left; Not the notebook

Here are a few ways you can interact with the graph.
* Click the empty space in the middle, and hit your **Space** key to add vertices.
* **Ctrl-A** to select all vertices (and their edges).
* **Alt-feft-mouse-drag** (might take a couple times, there's a bug) to rectangle select vertices.
* **Ctrl-X** to cut all selected vertices.
* **Ctrl-C** to copy all selected vertices.
* **Ctrl-V** to paste all selected vertices.
* **Ctrl-/** to connect all selected vertices (in a cycle, if more than two).
