The simplest graph you can make is an edge with two vertices. The graph.addVertex() function returns a `Vertex` object, which has an `id` property (as well as some others we'll get to). We call `graph.addVertex(...)` twice, and pass the return values' id properties to the `graph.addEdge(...)` method. 

To see the result, you might have to click `Minimize` above this notebook, and or rotate and drag the animation behind the notebook. Click `Run` in the code cell below the next code cell, then click Minimize (above the notebook) to see the result. Clicking `Expand` afterward will bring it back. 

```javascript
edge = graph.addEdge(
  graph.addVertex().id,
  graph.addVertex().id
)
```

Now that we have an edge object, we can manipulate its color. Let's set it to something random.

```javascript
edge.color = Math.floor(Math.random() * 0xffffff);
// mash the Run button
```

Here are all the properties you can change on a vertex. Note that the addVertex method accepts an object. You can specify any or all the properties for the vertex to be initialized with.

```javascript
graph.addVertex({
  color: 0x6495ed,
  selectionColor: 0xffa500,
  texture: 'https://picsum.photos/200/200/?random=' + Math.random(),
  size: 2.0,
  label: "Hello, World!",
  selected: true,
  visible: true
});
```

The label property is special. Above it accepts a string, below it accepts an object with a text property and css properties (in lower camelcase). Either is accepted.

```javascript
graph.addVertex({
  label: {
    text: "Hello, World!",
    fontSize: "72",
    color: 0xffffff
    backgroundColor: 0x000000
  }
})
```

Here are the edge properties.

```javascript
graph.addEdge(
  graph.addVertex({ size: 0.5 }).id,
  graph.addVertex({ size: 0.5 }).id,
  {
    arrow: true, // might have to make vertices smaller to see the arrow
    color: 'black', // css color string or hex color
    visible: false
  }
)
```
